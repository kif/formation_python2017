```python
def collatz(value):
    if value <= 1:
        return 0
    if value%2 == 0:
        dist = collatz(value/2)+1
    else:
        dist = collatz(3*value+1)+1
    return dist
```

```python
[collatz(i) for i in range(int(1e6))]
```

```python
def collatz_cache(value, cache={0:0, 1:0}):
    if value in cache:
        return cache[value]
    if value%2 == 0:
        dist = collatz_cache(value/2) + 1
    else:
        dist = collatz_cache(3*value+1) + 1
    cache[value] = dist
    return dist
```

```python
%time [collatz_cache(i) for i in range(int(1e6))]
```

```python
%time max([collatz(i) for i in range(int(1e6))])
%time max([collatz_cache(i) for i in range(int(1e6))])
```

```python
import numpy
numpy.all(numpy.array([collatz(i) for i in range(10000)])==numpy.array([collatz_cache(i) for i in range(10000)]))
```
